# Histogram and boxplot for continous 

```python
def his(x_axis,df):
    sns.histplot(x=x_axis,data=df)
    plt.show() 
```

```python
his('Age',df)
```

```python
his('Flight Distance',df)
```

```python
fig,ax=plt.subplots(1,2,sharey=True)

sns.boxplot(x='Departure Delay',data=df,ax=ax[0],showfliers=False)
sns.boxplot(x='Arrival Delay',data=df,ax=ax[1],showfliers=False)

plt.show()
```

# countplot for categories

```python
def count (x_axis):
    sns.countplot(x=x_axis,data=df)
    plt.show()
```
